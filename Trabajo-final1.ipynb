2. Funciones

```python
# 2.1
def agregar_producto(nombre, precio, cantidad):
    producto_existe = False
    for producto in inventario:
        if producto["nombre"] == nombre:
           producto_existe = True
           producto["cantidad"] = producto["cantidad"] + cantidad
    
    if producto_existe == False:
        inventario.append({"nombre":nombre, "precio":precio, "cantidad":cantidad})
    return          
```

```python
nombre_nuevo = input("¿Dime el nombre del producto?").lower()
precio_nuevo = int(input("pon el precio del producto"))
cantidad_nuevo = int(input("Dime la cantidad del producto")) 

agregar_producto(nombre_nuevo, precio_nuevo, cantidad_nuevo)

inventario
```

```python
# 2.2
def ver_inventario():
    for producto in inventario:
        print("Nombre:",producto["nombre"].capitalize(), ", Precio: $",producto["precio"], ", Cantidad:",producto["cantidad"] )
    return
```

```python
ver_inventario()
```

```python
# 2.3
def buscar_producto(nombre):
    for producto in inventario:
        if producto["nombre"] == nombre:
            print("Nombre:",producto["nombre"].capitalize(), ", Precio: $",producto["precio"], ", Cantidad:",producto["cantidad"])
```

```python
buscar_producto("diadema")
```

```python
#2.4
def actualizar_stock(nombre, cantidad):
    producto_existe = False
    for producto in inventario:
        if producto["nombre"] == nombre:
           producto_existe = True
           producto["cantidad"] = producto["cantidad"] + cantidad
    
    if producto_existe == False:
        print("Este producto no existe")
    return   
            
```

```python
actualizar_stock("diadema", 1)
```

```python
#2.5
def eliminar_producto(nombre):
    producto_existe = False
    for producto in inventario:
        if producto["nombre"] == nombre:
            producto_existe = True
            inventario.remove(producto)
    
    if producto_existe == False:
        print("Este producto no existe")
    return   
```

```python
eliminar_producto("diadema")
```

```python
# 2.6
def calcular_valor_inventario():
    valor_total = 0
    for productos in inventario:
        valor_total = (productos["precio"] * productos["cantidad"]) + valor_total
    print(valor_total,"euros")   
```

```python
calcular_valor_inventario()
```

```python
# 2.7
def realizar_compra():
    seguir_comprando = True
    carrito = []
    total_compra = 0
    while seguir_comprando == True:
        ver_inventario()
        selecciona_prenda = input("Por favor selecciona un producto o finaliza tu compra escribiendo finalizar").lower()
        if selecciona_prenda == "finalizar":
            seguir_comprando = False
        
        else:
            producto_existe = False
            for producto in inventario:            
                if producto["nombre"] == selecciona_prenda:
                    producto_existe = True
                    if producto["cantidad"] > 0:
                        carrito.append(selecciona_prenda)
                        producto["cantidad"] = producto["cantidad"] - 1                 
                        print("Has seleccionado",selecciona_prenda)
                        total_compra += producto["precio"]
                    else:
                        print("Sin stock, prueba con otro producto")    
                
            if producto_existe == False:
                print("Producto no encontrado, seleccione otro producto")

    print("Total de la compra:", total_compra, "€")
    return carrito, total_compra
    
```

```python
realizar_compra()
```
